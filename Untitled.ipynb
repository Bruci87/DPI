```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Carregar imagem existente no diretório
imagem = cv2.imread('placa_ruido1.png', 0)

# Verificar se a imagem foi carregada corretament
if imagem is None:
    print("Erro ao carregar a imagem. Verifique o caminho e o nome do arquivo.")
else:
    # Aplicar filtro de mediana
    imagem_filt_mediana = cv2.medianBlur(imagem, 5)

    # Mostrar imagem original e filtrada
    plt.subplot(1, 2, 1), plt.imshow(imagem, cmap='gray'), plt.title('Original')
    plt.subplot(1, 2, 2), plt.imshow(imagem_filt_mediana, cmap='gray'), plt.title('Filtro Mediana')
    plt.show()
```
